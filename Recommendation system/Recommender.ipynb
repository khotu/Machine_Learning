we will create a function for content based recommender system.

some features of DataFrame must have same name as function have :

    like User_Id :- 'userId'    
         Movies_Id :- 'movieId'         
        Rating     :- 'rating'        
        Geners     :- 'genres'
       Movie_name  :- 'title'

```python
def recommending_movies(movies_df, user_rating_df, userid):
    user_viewed_movies = user_rating_df[user_rating_df.userId == userid]
    user_viewed_movies_rating = movies_df[movies_df.movieId.isin(user_viewed_movies.movieId.tolist())]
    user_viewed_movies_rating['rating'] = user_viewed_movies.rating.tolist()

    
    movies_with_genre_df = pd.DataFrame()  
    for index, row in movies_df.iterrows():
        for genre in row["genres"]:
            movies_with_genre_df.at[index, genre] = 1
    movies_with_genre_df.fillna(value=0, inplace=True)

    
    user_movies_gener_table = pd.DataFrame()
    for index, row in user_viewed_movies_rating.iterrows():
        for genre in row["genres"]:
            user_movies_gener_table.at[index, genre] = 1
    user_movies_gener_table.fillna(value=0, inplace=True)
    user_movies_gener_table.reset_index(drop=True, inplace=True)
    
    
    user_interest_gener = user_movies_gener_table.T.dot(user_viewed_movies.rating)

    movies_gener_table = movies_with_genre_df.set_index(movies_df.movieId, drop=True)

    movie_recommended_value = ((movies_gener_table * user_interest_gener).sum(axis=1)) / user_interest_gener.sum()
    recommended_movies_priority = movie_recommended_value.sort_values(ascending=False)
    recommended_movies = movies_df[movies_df.movieId.isin(recommended_movies_priority.head(20).keys())]
    recommended_movies.reset_index(drop=True, inplace=True)
    recommended_movies_withid = recommended_movies.loc[:,['movieId', 'title']]
    return recommended_movies_withid
```

```python
print(recommending_movies(movies, rating, 1))
```
